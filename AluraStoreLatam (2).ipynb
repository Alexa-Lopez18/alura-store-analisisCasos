### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Facturación total por tienda

facturacion_t1 = tienda['Precio'].sum()
facturacion_t2 = tienda2['Precio'].sum()
facturacion_t3 = tienda3['Precio'].sum()
facturacion_t4 = tienda4['Precio'].sum()

print("Facturación Tienda 1:", facturacion_t1)
print("Facturación Tienda 2:", facturacion_t2)
print("Facturación Tienda 3:", facturacion_t3)
print("Facturación Tienda 4:", facturacion_t4)
```

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1','Tienda 2','Tienda 3','Tienda 4']
facturacion = [facturacion_t1,facturacion_t2,facturacion_t3,facturacion_t4]

plt.figure(figsize=(8,5))
plt.bar(tiendas,facturacion)

plt.title("Facturación total por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos")

plt.show()
```

# 2. Ventas por categoría

```python
categoria_t1 = tienda['Categoría del Producto'].value_counts()
categoria_t2 = tienda2['Categoría del Producto'].value_counts()
categoria_t3 = tienda3['Categoría del Producto'].value_counts()
categoria_t4 = tienda4['Categoría del Producto'].value_counts()

print("Categorías más vendidas Tienda 1")
print(categoria_t1)
```

```python
categoria_t1.head(10).plot(kind='bar')

plt.title("Categorías más vendidas - Tienda 1")
plt.xlabel("Categoría")
plt.ylabel("Cantidad vendida")

plt.show()
```

# 3. Calificación promedio de la tienda


```python
promedio_t1 = tienda['Calificación'].mean()
promedio_t2 = tienda2['Calificación'].mean()
promedio_t3 = tienda3['Calificación'].mean()
promedio_t4 = tienda4['Calificación'].mean()

print("Promedio Tienda 1:", promedio_t1)
print("Promedio Tienda 2:", promedio_t2)
print("Promedio Tienda 3:", promedio_t3)
print("Promedio Tienda 4:", promedio_t4)
```

```python
promedios = [promedio_t1,promedio_t2,promedio_t3,promedio_t4]

plt.figure(figsize=(8,5))
plt.bar(tiendas,promedios)

plt.title("Evaluación promedio por tienda")
plt.xlabel("Tienda")
plt.ylabel("Promedio")

plt.show()
```

# 4. Productos más y menos vendidos

```python
ventas_t1 = tienda['Producto'].value_counts()
ventas_t2 = tienda2['Producto'].value_counts()
ventas_t3 = tienda3['Producto'].value_counts()
ventas_t4 = tienda4['Producto'].value_counts()

print("Producto más vendido tienda 1:", ventas_t1.idxmax())
print("Producto menos vendido tienda 1:", ventas_t1.idxmin())

print("Producto más vendido tienda 2:", ventas_t2.idxmax())
print("Producto menos vendido tienda 2:", ventas_t2.idxmin())

print("Producto más vendido tienda 3:", ventas_t3.idxmax())
print("Producto menos vendido tienda 3:", ventas_t3.idxmin())

print("Producto más vendido tienda 4:", ventas_t4.idxmax())
print("Producto menos vendido tienda 4:", ventas_t4.idxmin())
```

# 5. Envío promedio por tienda

```python
envio_t1 = tienda['Costo de envío'].mean()
envio_t2 = tienda2['Costo de envío'].mean()
envio_t3 = tienda3['Costo de envío'].mean()
envio_t4 = tienda4['Costo de envío'].mean()

print("Costo envío promedio tienda 1:", envio_t1)
print("Costo envío promedio tienda 2:", envio_t2)
print("Costo envío promedio tienda 3:", envio_t3)
print("Costo envío promedio tienda 4:", envio_t4)
```

```python
envios = [envio_t1,envio_t2,envio_t3,envio_t4]

plt.figure(figsize=(8,5))
plt.bar(tiendas,envios)

plt.title("Costo promedio de envío por tienda")
plt.xlabel("Tienda")
plt.ylabel("Costo envío")

plt.show()
```
